#### Part 1 Initial Processing

```python
#Basic Imports
import numpy as np
import pandas as pd
```

```python
#Load the data
train=pd.read_csv('./input/train.csv')
train
```

```python
#This code will be called later
y=train['SalePrice']
```

```python
#Load the data
test=pd.read_csv('./input/test.csv')
test
```

```python
testid=test['Id'] #This will be used later to generate the submission file
```

```python
#Drop the Id column and SalePrice column in the training set. 
#The Id column will not be used in the model. And SalePrice will be the y for the training.
train=train.drop(['SalePrice'],axis=1)
train=train.drop(['Id'],axis=1)
train.head()
```

```python
#Drop the Id column in the test set
#The Id column will not be used in the model.
test=test.drop(['Id'],axis=1)
test.head()
```

```python
#Before handle NaN and dummy variables, concatenate the train and test set. 
#So that deal with NaN and dummy variables will be done together. 
alldata=pd.concat([train,test],axis=0)
alldata.head()
```

```python
#This code shows all numerical columns in the training set.
alldata._get_numeric_data().columns
```

```python
#For convenience I use this code to check how many null values are in these numerical (literally) columns
alldata[alldata._get_numeric_data().columns].isnull().sum()
```

```python
#Fill NaN using the mean of that column (by default ignored NaN)
alldata['LotFrontage']=alldata['LotFrontage'].fillna(alldata['LotFrontage'].mean())
alldata['MasVnrArea']=alldata['MasVnrArea'].fillna(alldata['MasVnrArea'].mean())
```

```python
#For these columns, NaN simply means '0' as this property does not have a basement at all
alldata['BsmtFinSF1']=alldata['BsmtFinSF1'].fillna(0)
alldata['BsmtFinSF2']=alldata['BsmtFinSF2'].fillna(0)
alldata['BsmtUnfSF']=alldata['BsmtUnfSF'].fillna(0)
alldata['TotalBsmtSF']=alldata['TotalBsmtSF'].fillna(0)
```

```python
#For these columns, NaN means 0 since there is no bathroom at the basement (or no basement at all)
alldata['BsmtFullBath']=alldata['BsmtFullBath'].fillna(0)
alldata['GarageArea']=alldata['GarageArea'].fillna(0)
```

```python
#Based on previous results, GarageYrBlt column has 159 NaN. 
#NaN simply indictes: there is no garage at this property.
#These NaN values will be filled using the 0 to reflect the fact that there is no garage
alldata['GarageYrBlt']=alldata['GarageYrBlt'].fillna(0)
```

```python
#For these columns, NaN means 0 since there is no garage
alldata['GarageCars']=alldata['GarageCars'].fillna(0)
alldata['BsmtHalfBath']=alldata['BsmtHalfBath'].fillna(0)
```

```python
#Now all NaN values in columns with numbers are filled. Now deal with NaN in other columns
#In these columns, NaN means 'None', as this property does not have something
#One function will get the job done. 
alldata=alldata.fillna('None')
alldata.head()
```

```python
#Check if there is still an NaN value
alldata.isnull().sum()
```

```python
stringcolumns=alldata.select_dtypes(include=['object']) #Get all categorical columns
stringcolumns #Check whether they are useful to the model later
```

```python
description=stringcolumns.describe() #Basic statistics of these categorical columns.
description #To read the full table I wrote a CSV, but that code is not here to avoid confusions.
```

```python
fewcategories=stringcolumns[['Street','Alley','Utilities','LandSlope','CentralAir','PavedDrive']]
fewcategories #Columns with less than or equal to 3 unique categories are selected 
```

```python
#The next 6 countplots will show whether a column is useful for the model
#If a column almost completely consists of 1 category according to the countplot it will be dropped.
import seaborn as sns
sns.countplot(x='Street', data=fewcategories) #This one will be kept as the other category is still visible.
```

```python
sns.countplot(x='Alley', data=fewcategories) #This one will be kept.
```

```python
sns.countplot(x='Utilities', data=fewcategories) #This one will be DROPPED.
```

```python
sns.countplot(x='LandSlope', data=fewcategories) #This one will be kept.
```

```python
sns.countplot(x='CentralAir', data=fewcategories) #This one will be kept.
```

```python
sns.countplot(x='PavedDrive', data=fewcategories) #This one will be kept. 
```

```python
alldata=alldata.drop(['Utilities'],axis=1) #Drop the the 'Utilities' column.
```

```python
alldata=pd.get_dummies(alldata)
alldata.head()
```

```python
train=alldata[:train.shape[0]] #Split train/test
train 
```

```python
test=alldata[train.shape[0]:]
test
```

```python
%matplotlib inline 
import seaborn as sns
sns.distplot(y) #Check the skewness of the distribution of train['SalePrice']
```

```python
import numpy as np #Log transformation of the target value ('SalePrice') in the train set
y=np.log1p(y)
sns.distplot(y)
```

```python
X_train=train
y_train=y #See part 1
```

```python
X_train
```

```python
y_train
```

#### Part 2 The Naive Model

```python
#This null simply predict that the price would be the mean of y_train for all properties
y_test_naivemodel=pd.Series([np.mean(y_train)] * (y_train.size-1))
y_test_naivemodel.head()
```

```python
y_test_naivemodel=np.expm1(y_test_naivemodel) #Convert logarithms back to dollars
y_test_naivemodel.head()
```

```python
np.count_nonzero(y_test_naivemodel) #Just check the shape
```

```python
#The performance will be evaluated by submitting a file to Kaggle 
#Kaggle uses RMSE calculated by logarithms of predicted prices and the real prices for performance evaluation
#The remaining codes in this part: create a Kaggle submission file
sub1_Naive=pd.DataFrame(testid)
sub1_Naive.head()
```

```python
sub2_Naive=pd.DataFrame(y_test_naivemodel,columns=['SalePrice'])
sub2_Naive.head()
```

```python
sub_Naive=pd.concat([sub1_Naive,sub2_Naive],axis=1)
sub_Naive.head()
```

```python
sub_Naive.to_csv('NaiveModel.csv',index=False)
```

**Kaggle Score 0.41901**
